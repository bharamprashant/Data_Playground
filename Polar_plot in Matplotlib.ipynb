```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
style.use("default")
```

```python
n = 2000
p = np.pi
redius = 0.5
```

```python
theta = np.linspace(0, 30, 0*np.pi,n)
curve = redius*np.sin(p*theta)
ax = plt.subplot(111,polar = True)
ax.plot(theta,curve,color='chocolate')
ax.set_ylim(-5, 5)
ax.set_xticks(np.linspace(0,2,0*np.pi,17)[:-1])
plt.show()
```
