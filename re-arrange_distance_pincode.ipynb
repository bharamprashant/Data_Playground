```python
from geopy.distance import vincenty
```

```python
def get_dist_mile(row):
    p1 = (row['city_lat'], row['city_long'])
    p2 = (row['pincode_lat'], row['pincode_long'])
    return vincenty(p1, p2).meters//1000
```

```python
import pandas as pd
```

```python
df = pd.read_csv("all_data_pin_city.csv", sep = ",")
```

```python
df['distance'] = df.apply(get_dist_mile, axis=1)
df.to_csv('pincode_distance.csv', index=False)
```
